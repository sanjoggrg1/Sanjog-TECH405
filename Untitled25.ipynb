```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load data
df = pd.read_csv('Titanic-Dataset.csv')
```

```python
# Basic info
print(df.head())
print(df.info())
print(df.describe())
```

```python
# Check missing values
print(df.isnull().sum())
```

```python
# Visualizations
sns.countplot(x='Survived', data=df)
plt.show()
```

```python
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.show()
```

```python
# Data cleaning
# Fill missing values
df['Age'].fillna(df['Age'].mean(), inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)

# Drop unnecessary columns
df.drop(['Name', 'Ticket', 'Cabin'], axis=1, inplace=True)

# Encode categorical data
df = pd.get_dummies(df, columns=['Sex', 'Embarked'], drop_first=True)

print(df.head())
```

```python
# Feature Selection
corr = df.corr()

# Target correlation
target_corr = corr['Survived'].sort_values(ascending=False)
print(target_corr)

# Select features with correlation > 0.1
selected_features = target_corr[abs(target_corr) > 0.1].index
selected_features = selected_features.drop('Survived')

print("Selected Features:", selected_features)

# Final dataset
X = df[selected_features].values
y = df['Survived'].values.reshape(-1,1)
```

```python
# Normalize Data
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
# Neural Network
input_size = X.shape[1]
hidden_size = 8
output_size = 1

# Initialize weights
np.random.seed(42)
W1 = np.random.randn(input_size, hidden_size)
b1 = np.zeros((1, hidden_size))

W2 = np.random.randn(hidden_size, output_size)
b2 = np.zeros((1, output_size))
```

```python
# Activation Function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    return x * (1 - x)
```

```python
# Forward Pass
def forward(X):
    global Z1, A1, Z2, A2

    Z1 = np.dot(X, W1) + b1
    A1 = sigmoid(Z1)

    Z2 = np.dot(A1, W2) + b2
    A2 = sigmoid(Z2)

    return A2
```

```python
# Loss Function
def compute_loss(y, y_pred):
    return np.mean((y - y_pred)**2)
```

```python
# Backpropogation
def backward(X, y, y_pred, lr=0.01):
    global W1, b1, W2, b2

    m = X.shape[0]

    # Output layer error
    dZ2 = (y_pred - y) * sigmoid_derivative(y_pred)
    dW2 = np.dot(A1.T, dZ2) / m
    db2 = np.sum(dZ2, axis=0, keepdims=True) / m

    # Hidden layer error
    dZ1 = np.dot(dZ2, W2.T) * sigmoid_derivative(A1)
    dW1 = np.dot(X.T, dZ1) / m
    db1 = np.sum(dZ1, axis=0, keepdims=True) / m

    # Update weights
    W1 -= lr * dW1
    b1 -= lr * db1
    W2 -= lr * dW2
    b2 -= lr * db2

```

```python
# Training Loop
epochs = 1000
losses = []

for i in range(epochs):
    y_pred = forward(X)
    loss = compute_loss(y, y_pred)
    losses.append(loss)

    backward(X, y, y_pred)

    if i % 100 == 0:
        print(f"Epoch {i}, Loss: {loss}")
```

```python
# Loss Visualization
plt.plot(losses)
plt.title("Loss Curve")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.show()
```

```python
# Predicition & Accuracy
predictions = forward(X)
predictions = (predictions > 0.5).astype(int)

accuracy = np.mean(predictions == y)
print("Accuracy:", accuracy)
```
